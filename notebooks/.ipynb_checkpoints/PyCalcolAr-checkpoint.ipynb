## Operazioni derivate dal file di calibrazione

Prima di tutto si calcolano i gain dei colettori (aggiungere equazione?)

```python
# calcolo gain F/IC0
value_36F = float(calibration_data['36F'].values)
value_36IC0 = float(calibration_data['36IC0'].values)
gain_F_IC0 = float(value_36F / value_36IC0)
print ('gain F/IC0: ', gain_F_IC0)
```

```python
# calcolo gain F/IC1 
value_36F = float(calibration_data['36F'].values)
value_36IC1 = float(calibration_data['36IC1'].values)
gain_F_IC1 = float(value_36F / value_36IC1)
print ('gain F/IC1: ', gain_F_IC1)
```

```python
# calcolo 36IC0 correzione gain
corr_gain_36IC0 = value_36IC0 * gain_F_IC0
print ('corr_gain_36IC0: ', corr_gain_36IC0) 
```

```python
# calcolo 40/36 correzione
value_40F = float(calibration_data['40F'].values)
B = 62415000 #aggiungere descrizione della constante
corr_40_36 = (value_40F * B) / corr_gain_36IC0
print ('corr_40_36: ', corr_40_36)
```

Definisco tutti i parametri delle tabelle (A) e (B) del file excel (non sappiamo ancora da dove vengono presi tutti)

```python
# ho introdotto una modifica non sostanziale che migliora la leggibilità del codice
background_spectrometer_dict = {'Background 40Ar': [0.0020, 0.1],
                                'Background 39Ar': [0.0030, 0.1],
                                'Background 38Ar': [0.0010, 0.1],
                                'Background 37Ar': [0.0040, 0.2],
                                'Background 36Ar': [0.0005, 0.1]}

background_spectrometer_df = pd.DataFrame.from_dict(background_spectrometer_dict, orient='index')
background_spectrometer_df.columns = ['value', 'relative error']

display(background_spectrometer_df)
```

```python
irradiantions_indexes = ['Sample weight [g]', 'J factor', 'Atmospheric Ratio', 'Gain F/IC0', 'Gain rel uncertainty', \
                         'Measured 40Ar/36Ar pipettes', 'Gain corrected 40Ar/36Ar pipettes', 'Pipette rel uncertainty', \
                        'Total fractionation uncertainty', 'Sensitivity (mL/mV)', '(36Ar/37Ar) Ca', '(38Ar/37Ar) Ca', \
                        '(39Ar/37Ar) Ca', 'Lambda Ar37 [1/d]', 'Lambda Ar40 [1/Ma]', 'Interference 40K', 'Coefficient 39Ar for J', \
                        'Coefficient Ca/K', 'Coefficient Cl/K']
irradiantions_values = [0.00034, 0.0011422, 298.56, 1.15893481706379, 0.00476533496490375/1.15893481706379, 335.310170019032, 335.310170019032/1.15893481706379, \
                       0.00405352150904286, 0.00405352150904286, 0.0000000000337, 0.000255, 0.00027, 0.00067, 0.019748, 0.0005543, \
                       0.033, 143, 1.94, 0.178]
irradiantions_values_errors = [np.nan, 0.0005, np.nan, np.nan, 0.00476533496490375, np.nan, 0.96907, np.nan, np.nan, np.nan, \
                               0.015, 0.015, 0.015, np.nan, np.nan, 0.01, np.nan, np.nan, np.nan]


irradiantions_df = pd.concat([pd.Series(irradiantions_values), pd.Series(irradiantions_values_errors)], axis = 1)
irradiantions_df.index = irradiantions_indexes
irradiantions_df.columns = ['value', 'relative error']

display(irradiantions_df)
```

###### Svolgimento operazioni presenti nel fogli excel CalcolAr

Le operazione necessarie verranno aggiunte in un unico dataframe di risultati simile a quello del file excel 


```python
input_data_df = pd.DataFrame()
input_data_df.loc[:, '40Ar F'] = df_data.loc[:, '40Ar F'].values * 1000
input_data_df.loc[:, 'err40Ar F'] = df_data.loc[:, 'err40Ar F'].values * 1000

input_data_df.loc[:, '39Ar F'] = df_data.loc[:, '39Ar F'].values * 1000
input_data_df.loc[:, 'err39Ar F'] = df_data.loc[:, 'err39Ar F'].values * 1000

input_data_df.loc[:, '38Ar F'] = df_data.loc[:, '40Ar F'].values * 1000
input_data_df.loc[:, 'err38Ar F'] = df_data.loc[:, 'err38Ar F'].values * 1000

input_data_df.loc[:, '37Ar F'] = df_data.loc[:, '37Ar IC0'].values * df_data.loc[:, 'gainF/IC0'].values * 1000
input_data_df.loc[:, 'err37Ar F'] = df_data.loc[:, 'err37Ar IC0'].values * df_data.loc[:, 'gainF/IC0'].values * 1000

input_data_df.loc[:, '36Ar F'] = df_data.loc[:, '36Ar IC0'].values * df_data.loc[:, 'gainF/IC0'].values * 1000
input_data_df.loc[:, 'err36Ar F'] = df_data.loc[:, 'err36Ar IC0'].values * df_data.loc[:, 'gainF/IC0'].values * 1000

display(input_data_df)
```

```python
results_data = input_data_df
results_data.loc[:,'40Ar BC'] = results_data.loc[:, '40Ar F'].values - (results_data.loc[:, '39Ar F'].values * \
                                       irradiantions_df.loc['Interference 40K', 'value']) - background_spectrometer_df.loc['Background 40Ar', 'value']

results_data.loc[:,'1sigma_abs40'] = (pow(results_data.loc[:, 'err40Ar F'].values, 2) + pow(background_spectrometer_df.loc['Background 40Ar', 'value'] * \
                                                                                            background_spectrometer_df.loc['Background 40Ar', 'relative error'], 2))**(1/2)

results_data.loc[:,'1sigma_rel40'] = results_data.loc[:,'1sigma_abs40'].values / results_data.loc[:,'40Ar BC'].values

results_data.loc[:,'39Ar BC'] = results_data.loc[:, '39Ar F'].values - background_spectrometer_df.loc['Background 39Ar', 'value']

results_data.loc[:,'1sigma_abs39'] = (pow(results_data.loc[:, 'err39Ar F'].values, 2) + pow(background_spectrometer_df.loc['Background 39Ar', 'value'] * \
                                                                                            background_spectrometer_df.loc['Background 39Ar', 'relative error'], 2))**(1/2)

results_data.loc[:,'1sigma_rel39'] = results_data.loc[:,'1sigma_abs39'].values / results_data.loc[:,'39Ar BC'].values

results_data.loc[:,'38Ar BC'] = results_data.loc[:, '38Ar F'].values - background_spectrometer_df.loc['Background 38Ar', 'value']

results_data.loc[:,'1sigma_abs38'] = (pow(results_data.loc[:, 'err38Ar F'].values, 2) + pow(background_spectrometer_df.loc['Background 38Ar', 'value'] * \
                                                                                            background_spectrometer_df.loc['Background 38Ar', 'relative error'], 2))**(1/2)

results_data.loc[:,'1sigma_rel38'] = results_data.loc[:,'1sigma_abs38'].values / results_data.loc[:,'38Ar BC'].values

results_data.loc[:,'37Ar BC'] = results_data.loc[:, '37Ar F'].values - background_spectrometer_df.loc['Background 37Ar', 'value']

results_data.loc[:,'1sigma_abs37'] = (pow(results_data.loc[:, 'err37Ar F'].values, 2) + pow(background_spectrometer_df.loc['Background 37Ar', 'value'] * \
                                                                                            background_spectrometer_df.loc['Background 37Ar', 'relative error'], 2))**(1/2)

results_data.loc[:,'1sigma_rel37'] = results_data.loc[:,'1sigma_abs37'].values / results_data.loc[:,'37Ar BC'].values

results_data.loc[:,'36Ar BC'] = results_data.loc[:, '36Ar F'].values - background_spectrometer_df.loc['Background 36Ar', 'value']

results_data.loc[:,'1sigma_abs36'] = (pow(results_data.loc[:, 'err36Ar F'].values, 2) + pow(background_spectrometer_df.loc['Background 36Ar', 'value'] * \
                                                                                            background_spectrometer_df.loc['Background 36Ar', 'relative error'], 2))**(1/2)

results_data.loc[:,'1sigma_rel36'] = results_data.loc[:,'1sigma_abs36'].values / results_data.loc[:,'36Ar BC'].values


display(results_data)
```

```python
results_data.loc[:,'Decay Factor'] = (irradiantions_df.loc['Lambda Ar37 [1/d]', 'value'] * 1 * math.exp(irradiantions_df.loc['Lambda Ar37 [1/d]', 'value'] * 246))/(1-math.exp((-1)* irradiantions_df.loc['Lambda Ar37 [1/d]', 'value'] * 1))
display(results_data)
```

```python
results_data.loc[:,'Mult 4amu'] = irradiantions_df.loc['Gain corrected 40Ar/36Ar pipettes', 'relative error']
results_data.loc[:,'Mult 2amu'] = (results_data.loc[:,'Mult 4amu'].values + 1) / 2
results_data.loc[:,'Mult 1amu'] = (results_data.loc[:,'Mult 4amu'].values + 3) / 4

display(results_data)
```

```python
results_data.loc[:,'Ar36tot'] = results_data.loc[:,'36Ar BC'].values * results_data.loc[:,'Mult 4amu'].values
results_data.loc[:,'1sigRel36tot'] = (pow(results_data.loc[:,'1sigma_rel36'].values, 2) + pow(irradiantions_df.loc['Total fractionation uncertainty', 'value'], 2))**(1/2)

results_data.loc[:,'Ar38tot'] = results_data.loc[:,'38Ar BC'].values * results_data.loc[:,'Mult 2amu'].values
results_data.loc[:,'1sigRel38tot'] = (pow(results_data.loc[:,'1sigma_rel38'].values, 2) + 0.25 * pow(irradiantions_df.loc['Total fractionation uncertainty', 'value'], 2))**(1/2)

results_data.loc[:,'Ar39tot'] = results_data.loc[:,'39Ar BC'].values * results_data.loc[:,'Mult 1amu'].values
results_data.loc[:,'1sigRel39tot'] = (pow(results_data.loc[:,'1sigma_rel39'].values, 2) + 0.0625 * pow(irradiantions_df.loc['Total fractionation uncertainty', 'value'], 2))**(1/2)
                                                                        
display(results_data)
```
