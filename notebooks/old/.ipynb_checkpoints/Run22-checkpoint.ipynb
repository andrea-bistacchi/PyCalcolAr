```python
# importo librerie utili
import pandas as pd
import math
import numpy as np
import datetime
```

```python
# creo la lista con i nomi delle colonne
colNamesInitData_22 = ['40Ar F', 'err40Ar F', '38Ar IC0', 'err38Ar IC0', '36Ar IC1', 'err36Ar IC1','38Ar F', 'err38Ar F','36Ar IC0', 'err36Ar IC0 ', '39Ar F', 'err39Ar F', '37Ar IC0', 'err37Ar IC0' , '35Cl IC1', 'err35Cl IC1', 'gainIC0/IC1', 'err gainIC0/IC1' , '40F/36IC1',  'err 40F/36IC1',  '40F/36IC0' ,'err 40F/36IC0', 'Run', 'Path']
colnames_numeric_22 = ['40Ar F', 'err40Ar F', '38Ar IC0', 'err38Ar IC0', '36Ar IC1', 'err36Ar IC1','38Ar F', 'err38Ar F','36Ar IC0', 'err36Ar IC0 ', '39Ar F', 'err39Ar F', '37Ar IC0', 'err37Ar IC0' , '35Cl IC1', 'err35Cl IC1', 'gainIC0/IC1', 'err gainIC0/IC1' , '40F/36IC1',  'err 40F/36IC1',  '40F/36IC0' ,'err 40F/36IC0']
# definisco il percorso del file run.22
filePath_22 = "../test_data/Run22.txt"
# importo il file run.22
#df_Data = pd.read_csv(filePath, header = None, names = colNamesInitData, sep = [" ", ";"])  OKKIO TAB
df_Data_22 = pd.read_csv(filePath_22, header = None, names = colNamesInitData_22)
# qui si eliminano le graffe, ma forse si può fare in un unico passaggio all'importazione, definendo
# anche legraffe come separatori di elenco e imponendo di considerare separatori consecutivi come uno solo
# nota che ho sovrascritto il dataframe usando lo stesso nome, non serve crearne più copie che si accumulano
# in memoria
df_Data_22 = df_Data_22.replace(["{", "}"], ["", ""], regex=True)

# qui si convertono tutte le colonne del dataframe a numeric (float64), ma sicuramente si può fare
# sopra all'importazione
for i in range(len(colnames_numeric_22)):
    df_Data_22[colnames_numeric_22[i]] = pd.to_numeric(df_Data_22[colnames_numeric_22[i]])

# qui aggiungere le colonne "mancanti"

# qui si visualizza il dataframe importato e pulito
df_Data_22

# va ripetuto per file run.4 e triplo36 -> df_Data_4 df_Data_36

# poi i tre dataframe devono essere concatenati a darne uno nuovo con tutte le colonne
# come esempio df_Data = pd.concat([df_Data_22, df_Data_4, df_Data_36], axis = 0)  -- asse 0 = righe

# riga temporanea per evitare errori sotto
df_Data = df_Data_22
df_Data
```

```python
# qui si splitta la colonna 'Run' in due colonne con none del run e data/ora
run_split = df_Data['Run'].str.split(" run on ")

# qui di formatta la colonna con nome del run (una Pandas SERIES) e le si da un nome
run_name = run_split.str[0]
run_name = run_name.replace(["'"], [""], regex=True)
run_name.name = 'Run_Name'

# idem per data, inoltre si converte il tipo in datetime64
dataora = run_split.str[1]
dataora.name = 'Date_Time'
dataora = pd.to_datetime(dataora)

# print(dataora.values)
# print(dataora.values[0])
# print(type(dataora.values[0]))
# print(dir(dataora.values[0]))

# qui si concatenano le nuove serie all'inizio del datframe con dati
df_Data = pd.concat([run_name, dataora, df_Data], axis = 1)

# qui si elimina la colonna non più usata
df_Data.drop('Run',axis=1, inplace=True)

# qui si visualizza il dataframe con le due nuove colonne
df_Data
```

```python
# esempio di filtro in base alla data

# data in base a cui filtro
value_to_check = pd.Timestamp(2020, 6, 23)


# applico il filtro
df_Data_filtered = df_Data[df_Data['Date_Time'] > value_to_check]

df_Data_filtered
```

```python
# esempio di filtro in base alla data

# date in base a cui filtro
value_min = pd.Timestamp(year=2020, month=6, day=22, hour=14)
value_max = pd.Timestamp(year=2020, month=6, day=23, hour=0)


# applico il filtro
df_Data_filtered = df_Data[(df_Data['Date_Time'] > value_min) & (df_Data['Date_Time'] < value_max)]

df_Data_filtered
```
